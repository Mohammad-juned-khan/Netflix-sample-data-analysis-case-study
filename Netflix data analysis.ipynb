# 2- Highest number of TV shows and movies released by years, Visualize result.

```python
# all the columns have data type as object. First convert date column to date datatype
#create new column Date_N which include release data.
data["Date_N"] = pd.to_datetime(data["Release_Date"])
data.head()
```

```python
#check data type of Date_N column
data["Date_N"].dtypes
```

```python
# Count number of TV Shows and Movies released by year
year_release_count = data["Date_N"].dt.year.value_counts()
year_release_count= year_release_count.sort_values(ascending=False)
year_release_count
```

```python
year_release_count.plot(kind="bar")
plt.xlabel("Years")
plt.ylabel("Count of Release")
plt.title("Count of release by years")
plt.show()
```
